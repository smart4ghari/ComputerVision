```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
import cv2
```

```python
# opencv directly converts images into numpy arrays
pic = cv2.imread("DATA/00-puppy.jpg")
```

```python
type(pic)
```

```python
pic.shape
```

```python
plt.imshow(pic)
```

```python
# MATPLOTLIB ---> RGB
# OPENCV -----> BGR
crt_img = cv2.cvtColor(pic,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(crt_img)
```

```python
gray_img = cv2.imread("DATA/00-puppy.jpg",cv2.IMREAD_GRAYSCALE)
```

```python
gray_img.shape
```

```python
plt.imshow(gray_img,cmap = 'gray')
```

```python
# resizing images
crt_img.shape
```

```python
new_img = cv2.resize(crt_img,(1000,400))
```

```python
plt.imshow(new_img)
```

```python
# Changing the images by its ratio
wid_ratio = 0.5 # 50%
hei_ratio = 0.5 # 50%
```

```python
new_img2 = cv2.resize(crt_img,(0,0),crt_img,wid_ratio,hei_ratio)
```

```python
plt.imshow(new_img2)
```

```python
new_img2.shape
```

```python
# How to flip the images
flip_img = cv2.flip(crt_img,0)
plt.imshow(flip_img)
```

```python
flip_img = cv2.flip(crt_img,1)
plt.imshow(flip_img)
```

```python
flip_img = cv2.flip(crt_img,-1)
plt.imshow(flip_img)
```
